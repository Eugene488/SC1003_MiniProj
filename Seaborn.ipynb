```python
import seaborn as sns
```

```python
File = open("records.csv", "r")

# Read all lines from the file and store them in a list
raw_data = File.readlines()
# header = next(File)
# Close the file
File.close()
```

```python
labels = raw_data[0][0:-1].split(',')
print(labels)
```

```python
data_of_students = raw_data[1:]
# print(data_of_students) 
```

```python
Students_dict_data = {}
stu_cnt = 0
list_of_Tutgrp = []
list_of_studentID = []
list_of_school = []
list_of_name = []
list_of_gender = []
list_of_CGPA = []


for stu_num in data_of_students:    
    student_data = data_of_students[stu_cnt][0:-1].split(",")
    cnt = 0
    for cat in student_data:
        # append each label to a list in order
        if cnt == 2:
            list_of_school.append(cat)
        elif cnt == 3:
            list_of_name.append(cat)
        elif cnt == 4:
            list_of_gender.append(cat)
        elif cnt == 5:
            list_of_CGPA.append(cat)
        # Students_dict_data.update({labels[cnt]: cat})
        cnt += 1
    stu_cnt += 1

print(Students_dict_data)
```

```python
# print(list_of_school)
# print(list_of_name)
# print(list_of_gender)
# print(list_of_CGPA)
# print(len(list_of_school))

unq_school = set(list_of_school)
unq_gender = set(list_of_gender)

#print(unq_school)
print(unq_gender)
```

```python
unq_gender_list = list(unq_gender)
total = 0
gender_count = []
for string in unq_gender_list:
    count = list_of_gender.count(string)
    gender_count.append(count)
    total += count
    print(f"{string}: {count}")

```

```python
import matplotlib.pyplot as plt
import csv
```

```python
plt.pie(gender_count, labels=unq_gender_list, autopct='%1.1f%%')
plt.title("Gender Distribution", fontsize=20)
plt.show()
```

```python
unq_school_list = list(unq_school)
total = 0
school_count = []
for string in unq_school_list:
    count = list_of_school.count(string)
    school_count.append(count)
    total += count
    print(f"{string}: {count}")
```

```python
plt.figure(figsize=(17, 6))
plt.bar(unq_school_list, school_count, color="g", width=0.7, label = "No of students")
plt.xlabel("School", fontsize=14)
plt.ylabel("No of students", fontsize=14)
plt.title("Distribution of students in each school", fontsize=20)
plt.legend()
plt.show()
```

```python
list_of_CGPA_float = [float(i) for i in list_of_CGPA]

plt.figure(figsize=(10, 6))
plt.boxplot(list_of_CGPA_float)
plt.title("Boxplot of CGPA", fontsize=20)
```
